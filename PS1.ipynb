```python
#2
import math
def F(x):
    if x == 1:
        return 1
    parent = math.ceil(x / 3)
    return F(parent) + 2 * x
def compute_F_list(arr):
    return [F(x) for x in arr]
print("输入0退出程序")
while True:
        user_input = input("请输入数字（用逗号分隔多个数字）: ")
        if user_input.strip() == '0':
            print("再见！")
            break
        numbers = [int(x.strip()) for x in user_input.split(',')]
        result = compute_F_list(numbers)
        print("输入:", numbers)
        print("输出:", result)
        break
```

```python
#3.1
def find_number_of_ways_10_dice_recursive(x):
    if x < 10 or x > 60:
        return 0
    def recursive_helper(dice_remaining, target_sum):
        if dice_remaining == 0:
            return 1 if target_sum == 0 else 0
        if dice_remaining == 1:
            return 1 if 1 <= target_sum <= 6 else 0
        min_possible = dice_remaining
        max_possible = 6 * dice_remaining
        if target_sum < min_possible or target_sum > max_possible:
            return 0
        total_ways = 0
        for k in range(1, 7):  
            if target_sum - k >= dice_remaining - 1: 
                total_ways += recursive_helper(dice_remaining - 1, target_sum - k)
        return total_ways
    return recursive_helper(10, x)
print("输入0退出程序")
while True:
        user_input = input("\n请输入总和 (10-60): ")
        if user_input.strip() == '0':
            print("再见！")
            break
        x = int(user_input)
        if 10 <= x <= 60:
            result = find_number_of_ways_10_dice_recursive(x)
            print(f"方法数: {result:,}")
        else:
            print("错误：总和必须在10-60之间！")
        break
```

```python
#4.1
import random
def Random_integer(N):
    return [random.randint(0, 10) for _ in range(N)]
arr = Random_integer(10)
print(f"生成的数组: {arr}")
#4.2
from itertools import combinations
total_sum = 0
n = len(arr)
for subset_size in range(1, n + 1):
    for subset in combinations(arr, subset_size):
        subset_avg = sum(subset) / subset_size
        total_sum += subset_avg
print("所有子集平均值之和:", total_sum)
```

```python
#4.3
import matplotlib.pyplot as plt
import numpy as np

def Sum_averages_theoretical(N):
    """
    使用理论公式计算Sum_averages
    使用浮点数避免整数溢出
    """
    # 对于均匀分布[0,10]的随机数组，期望总和为 5.5 * N
    expected_sum = 5.5 * N  # 0-10均匀分布的期望值是5.5
    # 使用浮点数指数计算
    return expected_sum * (2.0 ** (N - 1))

# 计算从N=1到100的Sum_averages
N_values = list(range(1, 101))
Total_sum_averages = []

print("使用理论公式计算中...")
print("公式: Sum_averages = 5.5 × N × 2^(N-1)")
print("-" * 50)

for N in N_values:
    # 使用理论公式计算，避免数值溢出
    result = Sum_averages_theoretical(N)
    Total_sum_averages.append(result)
    
    if N % 10 == 0:
        magnitude = np.log10(result) if result > 0 else 0
        print(f"N={N}: Sum_averages ≈ {result:.2e} (10^{magnitude:.1f})")

# 绘制结果
plt.figure(figsize=(15, 12))

# 1. 线性图（显示小N范围）
plt.subplot(2, 2, 1)
plt.plot(N_values[:30], Total_sum_averages[:30], 'b-', linewidth=2)
plt.xlabel('Array Size (N)')
plt.ylabel('Sum of Averages')
plt.title('Sum of Averages vs Array Size\n(N=1 to 30)')
plt.grid(True, alpha=0.3)

# 标记关键点
for N in [1, 5, 10, 15, 20, 25, 30]:
    value = Total_sum_averages[N-1]
    plt.plot(N, value, 'ro', markersize=4)
    plt.annotate(f'N={N}', (N, value), xytext=(5, 5), textcoords='offset points', fontsize=8)

# 2. 半对数图 (y轴对数)
plt.subplot(2, 2, 2)
plt.semilogy(N_values, Total_sum_averages, 'r-', linewidth=2)
plt.xlabel('Array Size (N)')
plt.ylabel('Sum of Averages (log scale)')
plt.title('Semi-log Plot\n(Perfect Exponential Growth)')
plt.grid(True, alpha=0.3)
```

```python
#5.1
import numpy as np
import random

def create_matrix(N, M):
    matrix = np.zeros((N, M), dtype=int)
    matrix[0, 0] = 1
    matrix[N-1, M-1] = 1
    for i in range(N):
        for j in range(M):
            if (i == 0 and j == 0) or (i == N-1 and j == M-1):
                continue
            matrix[i, j] = random.randint(0, 1)
    
    return matrix
N, M = 5, 6
matrix = create_matrix(N, M)
print(f"{N}×{M} 矩阵:")
print(matrix)
```

```python
#5.2
def Count_path(matrix):
    if not matrix or not matrix[0]:
        return 0
    rows, cols = len(matrix), len(matrix[0])
    if matrix[0][0] == 0 or matrix[rows-1][cols-1] == 0:
        return 0
    memo = {}
    def dfs(i, j):
        if i >= rows or j >= cols or matrix[i][j] == 0:
            return 0
        if i == rows - 1 and j == cols - 1:
            return 1
        if (i, j) in memo:
            return memo[(i, j)]
        paths = dfs(i + 1, j) + dfs(i, j + 1)
        memo[(i, j)] = paths
        return paths
    
    return dfs(0, 0)
def test_count_path(): 
    matrix2 = [
        [1, 0, 1],
        [1, 1, 1],
        [0, 1, 1]
    ]
    print("\n测试1 - 有障碍的3x3矩阵:")
    for row in matrix2:
        print(row)
    print("路径数量:", Count_path(matrix2))  
test_count_path()
```

```python
#5.3
import numpy as np

def Count_path_dp(matrix):
    """
    使用动态规划计算从左上角到右下角的路径数量
    0表示障碍，1表示可通过
    """
    if not matrix or not matrix[0]:
        return 0
    
    rows, cols = len(matrix), len(matrix[0])
    
    # 如果起点或终点是障碍
    if matrix[0][0] == 0 or matrix[rows-1][cols-1] == 0:
        return 0
    
    # 创建DP表
    dp = [[0] * cols for _ in range(rows)]
    dp[0][0] = 1  # 起点
    
    # 填充第一行
    for j in range(1, cols):
        dp[0][j] = dp[0][j-1] if matrix[0][j] == 1 else 0
    
    # 填充第一列
    for i in range(1, rows):
        dp[i][0] = dp[i-1][0] if matrix[i][0] == 1 else 0
    
    # 填充其余位置
    for i in range(1, rows):
        for j in range(1, cols):
            if matrix[i][j] == 1:
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
            else:
                dp[i][j] = 0
    
    return dp[rows-1][cols-1]

def create_random_matrix(N, M, seed=None):
    """
    创建N×M随机矩阵
    - 左上角和右下角固定为1
    - 其余位置随机填充0或1
    """
    if seed is not None:
        np.random.seed(seed)
    
    # 创建随机矩阵
    matrix = np.random.randint(0, 2, (N, M))
    
    # 固定左上角和右下角为1
    matrix[0, 0] = 1
    matrix[N-1, M-1] = 1
    
    return matrix.tolist()

def run_experiment(N=10, M=8, num_runs=1000):
    """
    运行多次实验并统计结果
    """
    print(f"开始实验: N={N}, M={M}, 运行次数={num_runs}")
    print("=" * 50)
    
    path_counts = []
    zero_path_runs = 0  # 记录路径数为0的次数
    
    for i in range(num_runs):
        # 创建随机矩阵（每次使用不同的种子）
        matrix = create_random_matrix(N, M, seed=i)
        
        # 计算路径数量
        paths = Count_path_dp(matrix)
        path_counts.append(paths)
        
        if paths == 0:
            zero_path_runs += 1
        
        # 显示进度
        if (i + 1) % 100 == 0:
            print(f"已完成 {i + 1}/{num_runs} 次运行")
    
    # 计算统计信息
    mean_paths = np.mean(path_counts)
    std_paths = np.std(path_counts)
    median_paths = np.median(path_counts)
    max_paths = np.max(path_counts)
    min_paths = np.min(path_counts)
    
    # 输出结果
    print("\n" + "=" * 50)
    print("实验结果统计:")
    print("=" * 50)
    print(f"平均路径数量: {mean_paths:.2f}")
    print(f"路径数量标准差: {std_paths:.2f}")
    print(f"路径数量中位数: {median_paths:.2f}")
    print(f"最大路径数量: {max_paths}")
    print(f"最小路径数量: {min_paths}")
    print(f"路径数为0的次数: {zero_path_runs} ({zero_path_runs/num_runs*100:.1f}%)")
    
    return path_counts, mean_paths

def analyze_distribution(path_counts):
    """分析路径数量的分布"""
    print(f"\n{'路径数量分布分析':-^50}")
    
    # 将路径数量分组
    bins = [0, 1, 10, 50, 100, 500, 1000, float('inf')]
    bin_labels = ['0', '1', '2-10', '11-50', '51-100', '101-500', '501-1000', '>1000']
    
    hist, _ = np.histogram(path_counts, bins=bins)
    
    print("路径数量分布:")
    for i in range(len(hist)):
        percentage = hist[i] / len(path_counts) * 100
        print(f"  {bin_labels[i]:>8}: {hist[i]:4d} 次 ({percentage:5.1f}%)")
    
    # 计算非零路径的平均值
    non_zero_paths = [p for p in path_counts if p > 0]
    if non_zero_paths:
        mean_non_zero = np.mean(non_zero_paths)
        print(f"\n非零路径的平均值: {mean_non_zero:.2f}")
        print(f"非零路径的比例: {len(non_zero_paths)/len(path_counts)*100:.1f}%")

def visualize_sample_matrices(N=10, M=8, num_samples=5):
    """可视化几个样本矩阵"""
    print(f"\n{'样本矩阵示例':-^50}")
    
    for i in range(num_samples):
        matrix = create_random_matrix(N, M, seed=i)
        paths = Count_path_dp(matrix)
        
        print(f"\n样本 {i+1} (路径数量: {paths}):")
        # 显示矩阵的前几行和前几列
        for row in matrix[:4]:  # 只显示前4行
            print("  " + " ".join(str(x) for x in row[:6]) + " ...")
        if N > 4:
            print("  ...")

# 运行主实验
if __name__ == "__main__":
    N, M = 10, 8
    num_runs = 1000
    
    # 运行实验
    path_counts, mean_paths = run_experiment(N, M, num_runs)
    
    # 分析分布
    analyze_distribution(path_counts)
    
    # 显示样本
    visualize_sample_matrices(N, M)
    
    # 理论分析
    print(f"\n{'理论分析':-^50}")
    total_cells = N * M
    variable_cells = total_cells - 2  # 减去两个固定为1的角落
    print(f"矩阵大小: {N} × {M} = {total_cells} 个单元格")
    print(f"随机填充的单元格: {variable_cells} 个")
    print(f"每个单元格为1的概率: 50%")
    print(f"实验次数: {num_runs}")
    print(f"\n最终结果: 平均路径数量 = {mean_paths:.2f}")
```
